# pycryptodome 主要加解密方法
* 對稱金鑰密碼: 在Crypto.Cipher裡，有 DES, 3DES, AES, RC4
* 非對稱金鑰密碼: 在Crypto.PublicKey裡，有 RSA, DSA
* Hash密碼: 在Crypto.Hash裡，有 MD5, SHA-1, SHA-128, SHA-256
* 隨機數: 在Crypto.Random裡
* 小工具: 在Crypto.Util裡
* 數位簽章: 在Crypto.Signature裡

# DES實作
* key size = 64 bits<p>
* block size = 64 bits<p>
* 說明文件 [link text](https://pycryptodome.readthedocs.io/en/latest/src/cipher/des.html)<p>
    
註: 在字串和bytes之間轉換可以使用encode()和decode()函數，不加參數，預設以utf-8編碼。<p>
*   encode(): 字串 => bytes
*   decode(): bytes => 字串



```
# DES ex1: 基本 DES 加解密
from Crypto.Cipher import DES

# 金鑰需64-bits，ECB mode不需要Initial Vector
des = DES.new(b'01234567', DES.MODE_ECB) 

text = b'BrCTF{HappyHack}'
print("待加密的明文:%s" %text)

cipher_text = des.encrypt(text)
print("加密後的密文:%s" %cipher_text)

plain_text = des.decrypt(cipher_text)
print("解密後的答案:%s" %plain_text)
```

```
# DES ex2: 標準 DES 加解密
# 加密做法: plaintext => encrypt => Base64 encode => cipher 
# 解密做法: cipher => Base64 decode => decrypt => plaintext
from Crypto.Cipher import DES
import base64

key = b'TNFSH'.ljust(8, b' ')      # 利用 ljust 來補足key size和block size
des = DES.new(key, DES.MODE_ECB)

# encrypt
plain_text = b'this_is_a_plain'.ljust(16, b' ')
text_enc = des.encrypt(plain_text)
text_enc_b64 = base64.b64encode(text_enc)
print(text_enc_b64.decode())       # string decode(): https://www.runoob.com/python/att-string-decode.html

# decrypt
msg_enc = base64.b64decode(text_enc_b64)
msg = des.decrypt(msg_enc)
print(msg.decode())
```

```
# DES ex3: 採用CBC mode，需要IV
from Crypto.Cipher import DES
import binascii
import base64

# encrypt
key = b'12345678'
des = DES.new(key, DES.MODE_CBC, iv=b'87654321')
plain_text = '臺南一中傲視群雄'.encode()
text_enc = des.encrypt(plain_text)
print(text_enc)                               # 以 Bytes 型式輸出
print(binascii.b2a_hex(text_enc))             # 以 HEX 型式輸出
print(base64.b64encode(text_enc))             # 以 Base64 型式輸出
print(base64.b64encode(text_enc).decode())    # 以 Base64字串 型式輸出

#decrypt
des2 = DES.new(key, iv=des.iv, mode=DES.MODE_CBC)
msg = des2.decrypt(text_enc)
print(msg.decode())                           # 輸出解密明文時，建議加上decode()將bytes轉成字串
```

# DES練習: 請將 DES ex1 程式改為如 DES ex2 標準做法


```
# DES ex1: 基本 DES 加解密
from Crypto.Cipher import DES

# 金鑰需64-bits，ECB mode不需要Initial Vector
key = b'kaw12ew'.ljust(8,b' ')
des = DES.new(key, DES.MODE_ECB) 

text = b'BrCTF{HappyHack}'.ljust(32,b' ')
text_enc = des.encrypt(text)
text_enc_b64 = base64.b64encode(text_enc)
print(text_enc)

print(text_enc_b64.decode())       # string decode(): https://www.runoob.com/python/att-string-decode.html

# decrypt
msg_enc = base64.b64decode(text_enc_b64)
msg = des.decrypt(msg_enc)
print(msg.decode())
```

# 3DES 實作
* 每組 key size = 64 bits(8 Bytes)，但3DES有三個key，若key size為16 Bytes，則key3 = key1。
* 說明文件 [link text](https://pycryptodome.readthedocs.io/en/latest/src/cipher/des3.html)

```
# 3DES ex1: 基本 Triple DES 加解密
from Crypto.Cipher import DES3

# encrypt
key = b'01234567abcdefgh'
des3 = DES3.new(key, DES3.MODE_ECB)
text = b'BrCTF{HappyHack}'
cipher_text = des3.encrypt(text)
print("加密後的密文:%s" %cipher_text)

# decrypt
plain_text = des3.decrypt(cipher_text)
print("解密後的答案:%s" %plain_text)
```

# AES 實作
* key size 可以為128, 192, 256 bits (16, 24, 32 Bytes)
* block size 固定為 128 bits(16 Bytes)
* 說明文件 [link text](https://pycryptodome.readthedocs.io/en/latest/src/cipher/aes.html)

```
# AES ex1: 基本 AES 加解密
from Crypto.Cipher import AES
import base64

# encrypt
key = b'Sixteen byte key'         # 若不足16個字元，請用ljust(16, b' ')補足空白
aes = AES.new(key, AES.MODE_ECB)
plain_text = b'TNFSH in Tainan'.ljust(16, b' ')
text_enc = aes.encrypt(plain_text)
print(base64.b64encode(text_enc).decode())

# decrypt
msg = aes.decrypt(text_enc)
print(msg.decode())
```
