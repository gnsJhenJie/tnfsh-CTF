## 求私鑰d

```
def ext_gcd(a, b):
  if b == 0: 
    return 1, 0 ,a
  x, y, g = ext_gcd(b, a%b)
  x, y = y, x-(a//b)*y
  return x,y,g

p =  40639 #999983
q = 999983
N = p*q
phi = (p-1)*(q-1)
e = 65537
print('public-key = ', e)
d, k, gcd = ext_gcd(e, phi)
print('private-key = ', d)
```

## 加密

```
m = 123456
print("plain = ", m)
c = pow(m, e, N)
print('cipher = ', c)
```

## 解密

```
r = pow(c, d, N)
print('message = ',r)
```
